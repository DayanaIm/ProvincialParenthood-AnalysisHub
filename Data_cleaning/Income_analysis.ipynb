```python
import pandas as pd
import os
```

```python
directory_path = '../Data/Income/'

csv_files = ['Canada_2010-2021_income_by_economic_family_type.csv', 
             'Alberta_2010-2021_income_by_economic_family_type.csv', 
             'BC_2010-2021_income_by_economic_family_type.csv',
             'Manitoba_2010-2021_income_by_economic_family_type.csv',
             'New_Brunswick_2010-2021_income_by_economic_family_type.csv',
             'Newfoundland_and_Lab_2010-2021_income_by_economic_family_type.csv',
             'Nova_Scotia_2010-2021_income_by_economic_family_type.csv',
             'Ontario_2010-2021_income_by_economic_family_type.csv',
             'PEI_2010-2021_income_by_economic_family_type.csv',
             'Quebec_2010-2021_income_by_economic_family_type.csv',
             'Saskatchewan_2010-2021_income_by_economic_family_type.csv']


full_paths = [os.path.join(directory_path, file) for file in csv_files]
```

```python
dfs = []

for csv_file in full_paths:
    df = pd.read_csv(csv_file)
    dfs.append(df)
    
combined_df = pd.concat(dfs, axis=0, ignore_index=True)

combined_df
```

```python
missing_values = combined_df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
combined_df = combined_df.drop(columns=['UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS'])
combined_df
```

```python
missing_values = combined_df.isnull().sum()
print("Missing Values:\n", missing_values)
```
